```python
df = pd.read_csv(r'C:\Users\noemi\Documents\GitHub\Data-mining-project\datasets_final\charts_with_features.csv')
```

```python
# Define the regions you want to check
regions_to_check = ['Latin America', 'Oceania', 'Western Europe', 'Eastern Europe', 
                    'North America', 'Northern Europe', 'Africa', 'Southern Europe', 
                    'East Asia', 'South Asia', 'Middle East']

# Filter the DataFrame to include only rows from the specified regions
df_filtered = df[df['region'].isin(regions_to_check)]

# Group by 'track_id' and 'region' to calculate the mean of 'rank' and the frequency (count of occurrences)
df_grouped = df_filtered.groupby(['track_id', 'region']).agg({
    'rank': 'mean',    # Mean of rank (new column will be 'rank_mean')
    'track_id': 'size' # Count of occurrences (this is the frequency)
}).rename(columns={'rank': 'rank_mean', 'track_id': 'frequency'}).reset_index()

# Drop duplicates for the 'track_id' and 'region', and keep the remaining columns
df_unchanged = df_filtered.drop(columns=['rank']).drop_duplicates(subset=['track_id', 'region'])

# Merge the grouped data with the remaining unchanged columns
df_final = pd.merge(df_grouped, df_unchanged, on=['track_id', 'region'], how='left')

# Save the final DataFrame to a CSV file
output_file = r'C:\Users\noemi\Documents\GitHub\Data-mining-project\datasets\Final_version\charts_with_features.csv'
df_final.to_csv(output_file, index=False)
```

```python
# Check for duplicates in 'id' and 'region'
duplicates = df_final[df_final.duplicated(subset=['track_id', 'region'], keep=False)]

if not duplicates.empty:
    print("Duplicate ids found for the same region:")
    print(duplicates)
else:
    print("No duplicates found. Each region has unique ids.")
```

```python
#Outlier detection

# List of features
features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 
            'acousticness', 'instrumentalness', 'liveness', 'valence', 
            'tempo', 'duration_ms', 'time_signature']

# Create box plots for each feature
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(4, 4, i)
    sns.boxplot(x=df_final[feature])
    plt.title(f'Box Plot of {feature}')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.spatial.distance import mahalanobis
from scipy.stats import chi2
```

```python
#Mahalanobis distance

# Subset of the data for the relevant features
data_features = df_final[features]

# Mean and covariance of the features
mean = np.mean(data_features, axis=0)
cov_matrix = np.cov(data_features.values.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)

# Function to compute Mahalanobis distance for each point
def mahalanobis_distance(row, mean, inv_cov_matrix):
    return mahalanobis(row, mean, inv_cov_matrix)

# Apply the function to each row of the dataset
data_features['mahalanobis_dist'] = data_features.apply(lambda row: mahalanobis_distance(row, mean, inv_cov_matrix), axis=1)

# Set threshold using chi-square distribution (95% confidence level)
threshold = chi2.ppf(0.95, df=data_features.shape[1]-1)  # df is the number of features

# Identify outliers (those with Mahalanobis distance greater than the threshold)
data_features['outlier'] = data_features['mahalanobis_dist'] > threshold

# Count the number of outliers
print(f"Number of outliers: {data_features['outlier'].sum()}")
```

```python
# Display the outliers
outliers = data_features[data_features['outlier']]
outliers
```

```python
#Delete the outliers
df_final = df_final[~data_features['outlier']]
df_final.head(5)
```

```python
#Decide if a song is popular or not based on the threshold of the rank and the frequency
```

```python
# Plotting histograms to analyze distribution
df['rank_mean'].hist(bins=50)
plt.title('Rank Mean Distribution')
plt.xlabel('Rank Mean')
plt.ylabel('Frequency')
plt.show()

df['frequency'].hist(bins=50)
plt.title('Frequency Distribution')
plt.xlabel('Frequency')
plt.ylabel('Count')
plt.show()

# Get descriptive statistics for rank_mean and frequency
print(df['rank_mean'].describe())
print(df['frequency'].describe())
```

```python
# Calculate percentiles for rank_mean and frequency
rank_threshold = df['rank_mean'].quantile(0.4)  # 30th percentile
frequency_threshold = df['frequency'].quantile(0.6)  # 75th percentile

print(f"Rank Threshold (30th percentile): {rank_threshold}")
print(f"Frequency Threshold (75th percentile): {frequency_threshold}")
```

```python
# Apply the thresholds
df['popular'] = 0
df.loc[(df['rank_mean'] < rank_threshold) & (df['frequency'] > frequency_threshold), 'popular'] = 1

# Check the distribution of popular vs non-popular songs
print(df['popular'].value_counts(normalize=True))
```

```python
# check distribution of each region 
df.groupby('region')['popular'].value_counts(normalize=True)
```

```python
# Create the 'popular' column based on the thresholds
df['popular'] = 0  # Initialize the column to 0 (not popular)

# Set 'popular' to 1 where the conditions are met
df.loc[(df['rank_mean'] < rank_threshold) & (df['frequency'] > frequency_threshold), 'popular'] = 1

df.to_csv(r'C:\Users\noemi\Documents\GitHub\Data-mining-project\datasets_final\charts_with_features.csv', index=False)
# Now 'df' has the new 'popular' column with values 1 or 0
```
