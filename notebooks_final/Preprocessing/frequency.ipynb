```python
#Show the distribution of the frequency
df['frequency'].value_counts()
```

```python
# Plot the distribution of the frequency
import matplotlib.pyplot as plt

df['frequency'].plot(kind='hist', bins=100, figsize=(12,6), facecolor='grey',edgecolor='black')
plt.title('Distribution of the frequency')
plt.xlabel('Frequency')
plt.ylabel('Number of songs')
plt.show()
```

```python
# Plot the histogram and zoom in on the lower frequency values
plt.figure(figsize=(10, 6))
plt.hist(df['frequency'], bins=100, color='gray', edgecolor='black')

# Set x-axis limit to zoom in on the beginning phase
plt.xlim(0, 500)  # Adjust this value as needed to zoom in more or less

plt.title('Distribution of the Frequency (Zoomed In)')
plt.xlabel('Frequency')
plt.ylabel('Number of Songs')
plt.show()
```
