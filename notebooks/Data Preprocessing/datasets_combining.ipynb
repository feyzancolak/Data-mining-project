```python
#optimization of the previous code
from datetime import timedelta
from multiprocessing import Pool, cpu_count

# Function to process each song id within a region
def process_id_group(group):
    # Sort by date
    group = group.sort_values(by='date')

    # Get the first date and one year later
    first_date = group['date'].min()
    one_year_later = first_date + timedelta(days=365)

    # Filter the data for the one-year period
    one_year_data = group[(group['date'] >= first_date) & (group['date'] <= one_year_later)]

    # Calculate the mean rank and sum of streams
    rank_mean = one_year_data['rank'].mean()
    streams_sum = one_year_data['streams'].sum()

    # Add the calculated values to the group
    one_year_data['rank_mean'] = rank_mean
    one_year_data['streams_sum'] = streams_sum

    return one_year_data

# Function to process each region in parallel
def process_region(region_data):
    # Group the data by 'id'
    grouped = region_data.groupby('id')

    # Process each id group
    result = pd.concat([process_id_group(group) for _, group in grouped])

    return result

# Function to parallelize the processing
def parallel_process(data, func):
    # Split the data by region
    regions = charts['region'].unique()
    data_by_region = [charts[charts['region'] == region] for region in regions]
    
    # Use a pool of workers to process each region in parallel
    with Pool(cpu_count()) as pool:
        results = pool.map(func, data_by_region)
    
    # Concatenate the results
    return pd.concat(results)

# Load your CSV file in chunks
chunksize = 10**6  # 1 million rows per chunk
result = pd.DataFrame()

for chunk in pd.read_csv(r'C:\Users\noemi\Documents\GitHub\Data-mining-project\datasets\Previous_versions\charts_filtered.csv', chunksize=chunksize, parse_dates=['date']):
    # Apply parallel processing for each chunk
    chunk_result = parallel_process(chunk, process_region)
    result = pd.concat([result, chunk_result])

# Save the final result to a new CSV file
result.to_csv(r'datasets\Final version\all_tracks_filtered.csv', index=False)

print("Processing complete. The new CSV file has been saved.")
```
