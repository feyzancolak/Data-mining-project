```python
#Authentication to the spotity API
import requests
import base64
import pandas as pd

#Spotify API credentials
client_id = 'client_id'
client_secret = 'client_secret'

def get_access_token(client_id, client_secret):
    auth_url = "https://accounts.spotify.com/api/token"
    auth_header = base64.b64encode(f"{client_id}:{client_secret}".encode()).decode()
    auth_data = {
        'grant_type': 'client_credentials'
    }
    auth_headers = {
        'Authorization': f'Basic {auth_header}',
        'Content-Type': 'application/x-www-form-urlencoded'
    }
    response = requests.post(auth_url, headers=auth_headers, data=auth_data)
    
    # Stampa il codice di stato e il corpo della risposta per la diagnosi
    print(f"Status Code: {response.status_code}")
    print(f"Response Text: {response.text}")
    
    # Prova a decodificare il JSON, se possibile
    try:
        response_data = response.json()
    except ValueError:  # includes simplejson.decoder.JSONDecodeError
        response_data = None
    
    if response_data and 'access_token' in response_data:
        return response_data['access_token']
    else:
        raise Exception("Failed to retrieve access token: " + (response_data.get('error', 'Unknown error') if response_data else 'No response data'))

def get_audio_features(track_id, access_token):
    header = {
        'Authorization': f'Bearer {access_token}'
    }

    print(header)

    url = f"https://api.spotify.com/v1/audio-features/{track_id}"

    response = requests.get(url, headers = header)
    
    if response.status_code == 429:
        retry_after = int(response.headers.get('Retry-After', 1))  # Default to 1 second if header not present
        print(f"Rate limited. Retrying after {retry_after} seconds...")
        print(f"Response Headers: {response.headers}")  # Stampa tutti gli header della risposta
        #time.sleep(retry_after)

    return response
# Prova a ottenere il token di accesso
try:
    access_token = get_access_token(client_id, client_secret)
    print("Access token obtained successfully.")
except Exception as e:
    print(e)


```
